```python
classifier = TextClassifier.load('sentiment-fast')
```

```python
# Function to get sentiment
def get_sentiment(text):
    # Create a Sentence object
    sentence = Sentence(text)
    # Predict sentiment
    classifier.predict(sentence)
    # Get sentiment label and score
    label = sentence.labels[0].value  # Sentiment label: POSITIVE or NEGATIVE
    score = sentence.labels[0].score  # Confidence score
    return label, score
```

TinyBert

```python
class YouTubeDataset(Dataset):
    def __init__(self, data, tokenizer, max_len, numerical_features):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.numerical_features = numerical_features

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        title = row['title']
        tags = row['tags']
        numerical_data = row[self.numerical_features].values.astype(float)  # Convert to float
        label = row['category_id']

        # Tokenize title and tags
        inputs = self.tokenizer(
            title + " " + tags,
            max_length=self.max_len,
            padding="max_length",
            truncation=True,
            return_tensors="pt",
        )

        return {
            'input_ids': inputs['input_ids'].squeeze(0),
            'attention_mask': inputs['attention_mask'].squeeze(0),
            'numerical_data': torch.tensor(numerical_data, dtype=torch.float),
            'label': torch.tensor(label, dtype=torch.long),
        }

# Tokenizer for TinyBERT
tinybert_tokenizer = AutoTokenizer.from_pretrained("huawei-noah/TinyBERT_General_4L_312D")
max_len = 128
train_dataset = YouTubeDataset(train_data, tinybert_tokenizer, max_len, numerical_features)
val_dataset = YouTubeDataset(val_data, tinybert_tokenizer, max_len, numerical_features)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)
```

```python
class TransformerWithStructuredData(nn.Module):
    def __init__(self, transformer_model_name, numerical_input_size, num_labels):
        super(TransformerWithStructuredData, self).__init__()
        # Load the TinyBERT model
        self.transformer = AutoModel.from_pretrained(transformer_model_name)
        transformer_hidden_size = self.transformer.config.hidden_size  # TinyBERT hidden size
        self.fc1 = nn.Linear(numerical_input_size, 64)
        self.fc2 = nn.Linear(transformer_hidden_size + 64, 128)
        self.fc_out = nn.Linear(128, num_labels)
        self.dropout = nn.Dropout(0.3)
        self.relu = nn.ReLU()

    def forward(self, input_ids, attention_mask, numerical_data):
        # Get transformer output
        transformer_output = self.transformer(input_ids=input_ids, attention_mask=attention_mask)

        # Extract the pooled [CLS] token representation
        pooled_output = transformer_output.last_hidden_state[:, 0, :]  # [CLS] token embedding

        # Process numerical data through a fully connected layer
        numerical_output = self.relu(self.fc1(numerical_data))

        # Combine the transformer output and numerical data
        combined = torch.cat((pooled_output, numerical_output), dim=1)

        # Further processing through FC layers
        x = self.relu(self.fc2(combined))
        x = self.dropout(x)
        x = self.fc_out(x)

        return x

# Model parameters
tinybert_model = TransformerWithStructuredData(
    transformer_model_name="huawei-noah/TinyBERT_General_4L_312D",  # TinyBERT model name
    numerical_input_size=len(numerical_features),
    num_labels=len(label_encoder.classes_),
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tinybert_model.to(device)
```

```python
def train_model(tinybert_model, train_loader, val_loader, epochs=3):
    optimizer = torch.optim.AdamW(tinybert_model.parameters(), lr=2e-5)
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        tinybert_model.train()
        total_loss = 0
        for batch in train_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            numerical_data = batch['numerical_data'].to(device)
            labels = batch['label'].to(device)

            optimizer.zero_grad()
            outputs = tinybert_model(input_ids, attention_mask, numerical_data)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        print(f"Epoch {epoch + 1}/{epochs}, Loss: {total_loss / len(train_loader)}")

        # Validation step
        tinybert_model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                numerical_data = batch['numerical_data'].to(device)
                labels = batch['label'].to(device)

                outputs = tinybert_model(input_ids, attention_mask, numerical_data)
                predictions = torch.argmax(outputs, dim=1)
                correct += (predictions == labels).sum().item()
                total += labels.size(0)

        accuracy = correct / total
        print(f"Validation Accuracy: {accuracy:.4f}")

# Train TinyBERT model
train_model(tinybert_model, train_loader, val_loader)
```

MiniLM

```python
from torch.utils.data import DataLoader, Dataset
from transformers import AutoTokenizer
import torch

class YouTubeDataset(Dataset):
    def __init__(self, data, tokenizer, max_len, numerical_features):
        self.data = data
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.numerical_features = numerical_features

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        title = row['title']
        tags = row['tags']
        numerical_data = row[self.numerical_features].values.astype(float)
        label = row['category_id']

        # Tokenize title and tags using MiniLM tokenizer
        inputs = self.tokenizer(
            title + " " + tags,
            max_length=self.max_len,
            padding="max_length",
            truncation=True,
            return_tensors="pt",
        )

        return {
            'input_ids': inputs['input_ids'].squeeze(0),
            'attention_mask': inputs['attention_mask'].squeeze(0),
            'numerical_data': torch.tensor(numerical_data, dtype=torch.float),
            'label': torch.tensor(label, dtype=torch.long),
        }

# Step 3: Initialize the tokenizer and dataset using MiniLM
tokenizer = AutoTokenizer.from_pretrained("microsoft/MiniLM-L12-H384-uncased")  # MiniLM tokenizer
max_len = 128

# Initialize datasets
train_dataset = YouTubeDataset(train_data, tokenizer, max_len, numerical_features)
val_dataset = YouTubeDataset(val_data, tokenizer, max_len, numerical_features)

# Initialize data loaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)
```

```python
class TransformerWithStructuredData(nn.Module):
    def __init__(self, transformer_model_name, numerical_input_size, num_labels):
        super(TransformerWithStructuredData, self).__init__()
        self.transformer = AutoModel.from_pretrained(transformer_model_name)
        transformer_hidden_size = 384  # MiniLM hidden size is 384
        self.fc1 = nn.Linear(numerical_input_size, 64)
        self.fc2 = nn.Linear(transformer_hidden_size + 64, 128)
        self.fc_out = nn.Linear(128, num_labels)
        self.dropout = nn.Dropout(0.3)
        self.relu = nn.ReLU()

    def forward(self, input_ids, attention_mask, numerical_data):
        transformer_output = self.transformer(
            input_ids=input_ids, attention_mask=attention_mask
        )
        pooled_output = transformer_output.last_hidden_state[:, 0, :]  # [CLS] token embedding

        numerical_output = self.relu(self.fc1(numerical_data))

        combined = torch.cat((pooled_output, numerical_output), dim=1)
        x = self.relu(self.fc2(combined))
        x = self.dropout(x)
        x = self.fc_out(x)
        return x

# Model parameters
minilm_model = TransformerWithStructuredData(
    transformer_model_name="microsoft/MiniLM-L12-H384-uncased",
    numerical_input_size=len(numerical_features),
    num_labels=len(label_encoder.classes_),
)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
minilm_model.to(device)
```

```python
optimizer = torch.optim.Adam(minilm_model.parameters(), lr=1e-5)
criterion = nn.CrossEntropyLoss()
```

```python
import matplotlib.pyplot as plt

def train_model(minilm_model, train_loader, val_loader, epochs=3):
    train_losses = []
    val_accuracies = []

    for epoch in range(epochs):
        minilm_model.train()
        total_loss = 0
        for batch in train_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            numerical_data = batch['numerical_data'].to(device)
            labels = batch['label'].to(device)

            optimizer.zero_grad()
            outputs = minilm_model(input_ids, attention_mask, numerical_data)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        avg_loss = total_loss / len(train_loader)
        train_losses.append(avg_loss)
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {avg_loss:.4f}")

        # Validation step
        minilm_model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for batch in val_loader:
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                numerical_data = batch['numerical_data'].to(device)
                labels = batch['label'].to(device)

                outputs = minilm_model(input_ids, attention_mask, numerical_data)
                predictions = torch.argmax(outputs, dim=1)
                correct += (predictions == labels).sum().item()
                total += labels.size(0)

        accuracy = correct / total
        val_accuracies.append(accuracy)
        print(f"Validation Accuracy: {accuracy:.4f}")

    # Visualization
    plt.figure(figsize=(10, 5))

    # Plotting Training Loss
    plt.subplot(1, 2, 1)
    plt.plot(range(1, epochs + 1), train_losses, marker='o', linestyle='-')
    plt.title('Training Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)

    # Plotting Validation Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(range(1, epochs + 1), val_accuracies, marker='o', linestyle='-', color='orange')
    plt.title('Validation Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Train the minilm_model
train_model(minilm_model, train_loader, val_loader)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

def evaluate_model(model, data_loader, model_name="Model"):
    model.eval()
    all_predictions = []
    all_labels = []
    with torch.no_grad():
        for batch in data_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            numerical_data = batch['numerical_data'].to(device)
            labels = batch['label'].to(device)

            outputs = model(input_ids, attention_mask, numerical_data)
            predictions = torch.argmax(outputs, dim=1)
            all_predictions.extend(predictions.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    print(f"Classification Report for {model_name}")
    print(classification_report(all_labels, all_predictions))

    acc = accuracy_score(all_labels, all_predictions)
    print(f"Accuracy for {model_name}: {acc:.4f}")

    return all_labels, all_predictions

# Evaluate both models
minilm_labels, minilm_predictions = evaluate_model(minilm_model, val_loader, model_name="MiniLM")
tinybert_labels, tinybert_predictions = evaluate_model(tinybert_model, val_loader, model_name="TinyBERT")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming data has been loaded into the `data` DataFrame

# Preprocessing steps already done (you provided these earlier)
data[['views', 'likes', 'dislikes']] = data[['views', 'likes', 'dislikes']].fillna(0).replace([np.inf, -np.inf], 0)
data['title'] = data['title'].astype(str)
data['tags'] = data['tags'].astype(str)

# Encoding category_id
label_encoder = LabelEncoder()
data['category_id'] = label_encoder.fit_transform(data['category_id'])

# Split into train and validation data
train_data, val_data = train_test_split(data, test_size=0.2, random_state=42)

# Scale numerical features
scaler = StandardScaler()
numerical_features = ['views', 'likes', 'dislikes', 'comment_count']
train_data[numerical_features] = scaler.fit_transform(train_data[numerical_features])
val_data[numerical_features] = scaler.transform(val_data[numerical_features])

# Combine title and tags into a single text column for TF-IDF extraction
train_data['combined_text'] = train_data['title'] + " " + train_data['tags']
val_data['combined_text'] = val_data['title'] + " " + val_data['tags']

# TF-IDF Vectorization for text data
tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # Limit number of features to 5000 for efficiency
X_train_text = tfidf_vectorizer.fit_transform(train_data['combined_text'])
X_val_text = tfidf_vectorizer.transform(val_data['combined_text'])

# Combine TF-IDF features with numerical features
X_train = np.hstack([X_train_text.toarray(), train_data[numerical_features].values])
X_val = np.hstack([X_val_text.toarray(), val_data[numerical_features].values])

# Train Logistic Regression model
lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train, train_data['category_id'])

# Make predictions on the validation set
lr_preds = lr_model.predict(X_val)

# Evaluate model performance
accuracy = accuracy_score(val_data['category_id'], lr_preds)
print(f"Logistic Regression Accuracy: {accuracy:.4f}")
print("\nClassification Report for Logistic Regression:")
print(classification_report(val_data['category_id'], lr_preds))

# Visualize the accuracy result
plt.bar(['Logistic Regression'], [accuracy], color=['blue'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Accuracy: Logistic Regression')
plt.show()
```
